**Hierarchial_Clustering**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage 
from matplotlib import pyplot as plt
```

```python
points=np.array([[2,2],[5,2],[7,3],[8,4],[10,6],[12,8],[5,5],[6,6],[5,8],[2,6],[3,7]])
```

```python
points
```

```python
labels = range(1, 12) 
plt.figure(figsize=(10, 7)) 
plt.subplots_adjust(bottom=0.1) 
plt.scatter(points[:,0],points[:,1], label='True Position')
```

```python
for label, x, y in zip(labels, points[:, 0], points[:, 1]): 
    plt.annotate(
    label,
    xy=(x, y), xytext=(-3, 3),
    textcoords='offset points', ha='right', va='bottom')
plt.show()
```

**Single Link Clustering**

```python
linked = linkage(points, 'single')
labellist=range(1,12)
dendrogram(linked, 
orientation='top',
labels=labellist,distance_sort='descending',show_leaf_counts=True)
plt.show()
```

**Complete Link Clustering**

```python
linked = linkage(points, 'complete')
labellist=range(1,12)
dendrogram(linked, 
orientation='top',
labels=labellist,distance_sort='descending',show_leaf_counts=True)
plt.show()
```

**Average Link Clustering**

```python
linked = linkage(points, 'average')
labellist=range(1,12)
dendrogram(linked, 
orientation='top',
labels=labellist,distance_sort='descending',show_leaf_counts=True)
plt.show()
```
